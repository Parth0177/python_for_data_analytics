### REAL WORLD SCENERIOS OF LIST

### create a to do list
take track of all tasks

```python
to_do_list=["wake up at 5","have breakfast","study python","go to gym"]
# adding a task
to_do_list.append("take protein")
to_do_list.append("go to market")
print(to_do_list)
```

```python
## remove the task
to_do_list.remove("take protein")
print(to_do_list)
```

```python
# checking if a task is in the list
if "take protein" in to_do_list:
  print("hey take protein")
else:
  print("you have completed the task")
```
