```python
### Array can be created using python module or using numpy python
```

```python
### creating array using python module
import array as arr

val= arr.array('i',[20,30,50,40,89,46,32])
print(val)
print(val[0])
print(val[1])
```

```python
for i in range(len(val)):
  print(val[i], end=" ")
```

```python
for i in range(len(val)//2):
  print(val[i], end=" ")
```

```python
### enhanced for loop
for i in val:
  print(i, end=" ")
```

```python
from array import *

vals=array('i',[20,30,50,40,89,46,32])
print(vals, end=",")
```

```python
for j in vals:
  print(j,end=".")
```

```python
print(vals.typecode)
```

```python
### reverse a array
vals.reverse()
for i in vals:
  print(i, end=" ")
```

```python
vals.insert(2, 100)
for i in vals:
  print(i, end=" ")
```

```python
vals.append(300)
for i in vals:
  print(i, end=" , ")
```

```python
vals[5]=10003
for i in vals:
  print(i,end=" , ")
```

```python
### copy array elemnts from one to other
copyArray = arr.array(vals.typecode, (i+1*8 for i in vals))
print(copyArray)
```

```python
for i in range(len(copyArray)):
  print(copyArray[i], end=" , ")
```

```python
copyArray.pop(5)
for i in range(len(copyArray)):
  print(copyArray[i], end=" , ")
```
