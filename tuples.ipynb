### TUPLES ARE IMMUTABLE AND ARE ORDERED COLLECTIONS OF ITEMS. SIMILAR TO LISTS BUT ARE IMMUTABLE


```python
### Creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
### Creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
numbers= tuple([1,2,3,4,5])
print(numbers)
```

```python
mixed_tuple= (1, "Hello", 3.14, [1,2,3], (4,5))
print(mixed_tuple)
```

```python
### Accessing tuple elements

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
print(numbers[len(numbers)//2])
```

```python
### Tuple Operations

numbers
```

```python
concatination_tuple=numbers+ mixed_tuple
print(concatination_tuple)
```

```python
numbers*3
```

```python
mixed_tuple*4
```
