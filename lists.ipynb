### list comprehension

```python
lst1=[]
for i in range(20):
  lst1.append(i**2)

print(lst1)
```

```python
[x**2 for x in range(10)] #list comprehension
```

```python
### Basic List Comprehension

square=[x**2 for x in range(10)]
print(square)
```

```python
### list comprehension with condition
even=[x for x in range(40) if x%2==0]
print(even)
```

```python
### Nested List Comprehension
lstt=[1,2,3,4]
lst2=['a','b','c','d']
pair=[(x,y)for x in lstt for y in lst2]
print(pair)
```

```python
### list comprehension with function calls

words=['hello','world','python','programming']
lengths=[len(word) for word in words]
print(lengths)
```

### ASSIGNMENTS

```python
### CREATE A LIST OF 20 POSITIVE INTEGERS
integers=[x for x in range(1,21)]
print(integers)
```

```python
### print first middle and last elements of the list

print(integers[0])
print(integers[len(integers)//2])
print(integers[-1])
```

```python
###print first 5 , last 5 and emelents between 5-15
print(integers[:5])
print(integers[15:])
print(integers[5:15])
```

```python
### using list comprehension print sqr of first 10 positive integers
sq=[x**2 for x in range(10)]
print(sq)
```

```python
### using list comprehension print only even numbers from list 1
even=[x for x in integers if x%2==0]
print(even)
```
