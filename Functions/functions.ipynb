```python
# why functions
num=24
if num%2==0:
  print("even")
else:
  print("odd")
```

```python
def even_or_odd(num):
  """This function determines if a number is even or odd."""
  if num%2==0:
    return "even"
  else:
    return "odd"
  
even_or_odd(24)
```

```python
even_or_odd(25)
```

```python
### function with multiple parameters
def add(num1, num2):
  """this add two numbers"""
  return num1+num2
```

```python
add(5,8)
```

```python
### default parameters
def greet(name="Guest"):
  """this function greets a person with hello"""
  print(f"Hello {name}! , I Miss You")
```

```python
greet("Parth")
```

```python
greet("Akshitaa")
```

```python
greet()
```

```python
###variable length arguments
### Positional and Keywords Arguments

def print_numbers(*args):
  for numbers in args:
    print(numbers)
```
